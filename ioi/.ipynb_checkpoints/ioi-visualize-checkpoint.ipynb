```python
from matplotlib import pyplot as plt
import numpy as np
import yaml
import torch
from os.path import join

import sys
sys.path.append('/home/leiyu/projects/def-yangxu/leiyu/circuit-discovery')
from dmc.circuit_gpt import CircuitGPT, CircuitGPTConfig

mask_dir = '/home/leiyu/scratch/circuit-discovery/mask_logits/'

def draw_line(ax, x1, y1, x2, y2, color, ls):
    if x1 == x2:
        y = np.linspace(y1, y2, 100)
        x = x1 + (y1-y)**2 - (y1-y2)*(y1-y) 
        ax.plot(x, y, ls, color=color, alpha=0.3)
    else:
        ax.plot([x1,x2], [y1,y2], color=color, ls=ls, alpha=0.3)
```

```python
def head_ablation(mask_logits_dict_edge, layer_is, head_js):
    ablated_mask_logits_dict_edge = {k:torch.clone(v) for k,v in mask_logits_dict_edge.items()}
    for layer_i, head_j in zip(layer_is, head_js):
        for k in range(layer_i, 12):
            mlp_masks_logits_k = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_mlp_logits']
            mlp_masks_logits_ki = mlp_masks_logits_k[1:][layer_i*13:(1+layer_i)*13]
            mlp_masks_logits_ki[head_j] = -1
    
            if k > layer_i:
                for l in range(12):
                    attn_q_masks_logits_kli = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_attention_q_logits'][:,l][1:][layer_i*13:(layer_i+1)*13]
                    attn_k_masks_logits_kli = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_attention_k_logits'][:,l][1:][layer_i*13:(layer_i+1)*13]
                    attn_v_masks_logits_kli = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_attention_v_logits'][:,l][1:][layer_i*13:(layer_i+1)*13]
                    
                    # print(attn_q_masks_logits_kli.shape)
                    
                    attn_q_masks_logits_kli[head_j] = -1.
                    attn_k_masks_logits_kli[head_j] = -1.
                    attn_v_masks_logits_kli[head_j] = -1.
    
        ablated_mask_logits_dict_edge['edge_mask_output_logits'][1:][layer_i*13:(1+layer_i)*13][head_j] = -1.
            
    return ablated_mask_logits_dict_edge
```

```python
mask_logits_dict_edge = torch.load(
    join(mask_dir, f'mask_logits_dict_edge_ioi_0_weight_257_edge_11.pt')
)

mask_logits_dict_edge = {k:v.detach().cpu() for k,v in mask_logits_dict_edge.items()}

l = 3
layer_is = torch.repeat_interleave(torch.arange(l),12)
head_js = torch.arange(12).repeat(l)

mask_logits_dict_edge = head_ablation(mask_logits_dict_edge, layer_is, head_js)
```

```python
output_mask = torch.where(mask_logits_dict_edge['edge_mask_output_logits'] > 0., 1., 0.)
attn_q_masks, attn_k_masks, attn_v_masks = [], [], []
mlp_masks = []

for i in range(12):
    attn_q_masks_logits_i = mask_logits_dict_edge[f'blocks.{i}.edge_mask_attention_q_logits']
    attn_k_masks_logits_i = mask_logits_dict_edge[f'blocks.{i}.edge_mask_attention_k_logits']
    attn_v_masks_logits_i = mask_logits_dict_edge[f'blocks.{i}.edge_mask_attention_v_logits']
    mlp_masks_logits_i = mask_logits_dict_edge[f'blocks.{i}.edge_mask_mlp_logits']

    attn_q_masks.append(torch.where(attn_q_masks_logits_i > 0., 1., 0.))
    attn_k_masks.append(torch.where(attn_k_masks_logits_i > 0., 1., 0.))
    attn_v_masks.append(torch.where(attn_v_masks_logits_i > 0., 1., 0.))
    mlp_masks.append(torch.where(mlp_masks_logits_i > 0., 1., 0.))
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 18))
ax.axis('off')

input_node_x, input_node_y = np.array([0.7, -0.2])

attn_q_nodes_x, attn_q_nodes_y = np.meshgrid(np.linspace(0, 1.4, 12)-0.03, np.linspace(0, 1.4, 12))
attn_k_nodes_x, attn_k_nodes_y = np.meshgrid(np.linspace(0, 1.4, 12), np.linspace(0, 1.4, 12))
attn_v_nodes_x, attn_v_nodes_y = np.meshgrid(np.linspace(0, 1.4, 12)+0.03, np.linspace(0, 1.4, 12))
attn_nodes_x, attn_nodes_y = np.meshgrid(np.linspace(0, 1.4, 12), np.linspace(0, 1.4, 12)+0.02)

mlp_nodes_x, mlp_nodes_y = np.meshgrid(np.array(0.7), np.linspace(0.07, 1.47, 12))
output_node_x, output_node_y = np.array([0.7, 1.6])
    
ax.scatter(input_node_x, input_node_y, color='grey')
ax.scatter(attn_q_nodes_x, attn_q_nodes_y, color='brown')
ax.scatter(attn_k_nodes_x, attn_k_nodes_y, color='red')
ax.scatter(attn_v_nodes_x, attn_v_nodes_y, color='orange')
ax.scatter(attn_nodes_x, attn_nodes_y, color='deeppink')

ax.scatter(mlp_nodes_x, mlp_nodes_y, color='green')
ax.scatter(output_node_x, output_node_y, color='blue')

n_edge_valid = 0
out_edge_dict = {}

for i in range(12):
    out_edge_dict[f'mlp-{i}'] = []
    for j in range(12):
        out_edge_dict[f'attn-{i}-{j}'] = []


###########################################################
# add edge connections for output node
if output_mask[0] == 1:
    draw_line(ax, input_node_x, input_node_y, output_node_x, output_node_y, color="blue", ls='--')
    n_edge_valid += 1
for i in range(12):
    output_mask_i = output_mask[1:][i*13:(1+i)*13]
    mlp_i_output_edge = output_mask_i[-1]
    if mlp_i_output_edge == 1:
        draw_line(ax, mlp_nodes_x[i][0], mlp_nodes_y[i][0], output_node_x, output_node_y, color="blue", ls='--')
        n_edge_valid += 1
        out_edge_dict[f'mlp-{i}'].append('output-node')
    for j in range(12):
        attn_ij_output_edge = output_mask_i[j]
        if attn_ij_output_edge == 1:
            draw_line(ax, attn_nodes_x[i,j], attn_nodes_y[i,j], output_node_x, output_node_y, color="blue", ls='--')
            n_edge_valid += 1
            out_edge_dict[f'attn-{i}-{j}'].append('output-node')


# add edge connections for mlp nodes
for i in range(12):
    edge_mask_mlps_i = mlp_masks[i]
    if edge_mask_mlps_i[0] == 1:
        draw_line(ax, input_node_x, input_node_y, mlp_nodes_x[i][0], mlp_nodes_y[i][0], color="green", ls='--')
        n_edge_valid += 1        
    for j in range(i+1):
        edge_mask_mlps_ij = edge_mask_mlps_i[1:][j*13:(1+j)*13]
        if j < i:
            edge_mask_mlps_i_mlp_j = edge_mask_mlps_ij[-1]
            if edge_mask_mlps_i_mlp_j == 1:
                draw_line(ax, mlp_nodes_x[i][0], mlp_nodes_y[i][0], mlp_nodes_x[j][0], mlp_nodes_y[j][0], color="green", ls='--')
                n_edge_valid += 1
                out_edge_dict[f'mlp-{i}'].append(f'mlp-{j}')
        for k in range(12):
            edge_mask_mlps_i_attn_jk = edge_mask_mlps_ij[k]
            if edge_mask_mlps_i_attn_jk == 1:
                draw_line(ax, mlp_nodes_x[i][0], mlp_nodes_y[i][0], attn_nodes_x[j,k], attn_nodes_y[j,k], color="green", ls='--')
                n_edge_valid += 1
                out_edge_dict[f'attn-{j}-{k}'].append(f'mlp-{i}')


# add edge connections for attn and  attn_qkv nodes
for i in range(12):
    for j in range(12):
        # the j-th attn head on the i-th layer

        if attn_q_masks[i][0, j] == 1:
            draw_line(ax, input_node_x, input_node_y, attn_q_nodes_x[i,j], attn_q_nodes_y[i,j], color="red", ls='--')
            n_edge_valid += 1
        if attn_k_masks[i][0, j] == 1:
            draw_line(ax, input_node_x, input_node_y, attn_k_nodes_x[i,j], attn_k_nodes_y[i,j], color="red", ls='--')
            n_edge_valid += 1
        if attn_v_masks[i][0, j] == 1:
            draw_line(ax, input_node_x, input_node_y, attn_v_nodes_x[i,j], attn_v_nodes_y[i,j], color="red", ls='--')
            n_edge_valid += 1
            
        for l in range(i):
            edge_mask_attn_q_ij_l = attn_q_masks[i][:, j][1:][l*13:(1+l)*13]
            edge_mask_attn_k_ij_l = attn_k_masks[i][:, j][1:][l*13:(1+l)*13]
            edge_mask_attn_v_ij_l = attn_v_masks[i][:, j][1:][l*13:(1+l)*13]

            if edge_mask_attn_q_ij_l[-1] == 1: # (mlp-l, attn_q_ij) 
                draw_line(ax, mlp_nodes_x[l][0], mlp_nodes_y[l][0], attn_q_nodes_x[i,j], attn_q_nodes_y[i,j], color="red", ls='--')
                n_edge_valid += 1
                out_edge_dict[f'mlp-{l}'].append(f'attn-{i}-{j}_q')
            if edge_mask_attn_k_ij_l[-1] == 1: # (mlp-l, attn_k_ij) 
                draw_line(ax, mlp_nodes_x[l][0], mlp_nodes_y[l][0], attn_k_nodes_x[i,j], attn_k_nodes_y[i,j], color="red", ls='--')
                n_edge_valid += 1
                out_edge_dict[f'mlp-{l}'].append(f'attn-{i}-{j}_k')
            if edge_mask_attn_v_ij_l[-1] == 1: # (mlp-l, attn_v_ij) 
                draw_line(ax, mlp_nodes_x[l][0], mlp_nodes_y[l][0], attn_v_nodes_x[i,j], attn_v_nodes_y[i,j], color="red", ls='--')
                n_edge_valid += 1
                out_edge_dict[f'mlp-{l}'].append(f'attn-{i}-{j}_v')

            for r in range(12):  # for each r-th (r = 1 to 12) attn head on the l-th layer (l < i) 
                if attn_q_masks[i][:, j][1:][l*13:(1+l)*13][r] == 1:
                    draw_line(ax, attn_nodes_x[l,r], attn_nodes_y[l,r], attn_q_nodes_x[i,j], attn_q_nodes_x[i,j], color="red", ls='--')
                    n_edge_valid += 1
                    out_edge_dict[f'attn-{l}-{r}'].append(f'attn-{i}-{j}_q')
                if attn_k_masks[i][:, j][1:][l*13:(1+l)*13][r] == 1:
                    draw_line(ax, attn_nodes_x[l,r], attn_nodes_y[l,r], attn_k_nodes_x[i,j], attn_k_nodes_x[i,j], color="red", ls='--')
                    n_edge_valid += 1
                    out_edge_dict[f'attn-{l}-{r}'].append(f'attn-{i}-{j}_k')
                if attn_v_masks[i][:, j][1:][l*13:(1+l)*13][r] == 1:
                    draw_line(ax, attn_nodes_x[l,r], attn_nodes_y[l,r], attn_v_nodes_x[i,j], attn_v_nodes_x[i,j], color="red", ls='--')
                    n_edge_valid += 1
                    out_edge_dict[f'attn-{l}-{r}'].append(f'attn-{i}-{j}_v')


plt.show()
# plt.savefig('gpt2-small-circuit-anaphor_gender_agreement.pdf')
 
```
