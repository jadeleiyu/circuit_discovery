```python
def head_ablation(mask_logits_dict_edge, layer_is, head_js):
    ablated_mask_logits_dict_edge = {k:torch.clone(v) for k,v in mask_logits_dict_edge.items()}

    for layer_i, head_j in zip(layer_is, head_js):
        for k in range(layer_i, 12):
            mlp_masks_logits_k = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_mlp_logits']
            mlp_masks_logits_ki = mlp_masks_logits_k[1:][layer_i*13:(1+layer_i)*13]
            mlp_masks_logits_ki[head_j] = -100.
    
            if k > layer_i:
                for l in range(12):
                    attn_q_masks_logits_kli = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_attention_q_logits'][:,l][1:][layer_i*13:(layer_i+1)*13]
                    attn_k_masks_logits_kli = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_attention_k_logits'][:,l][1:][layer_i*13:(layer_i+1)*13]
                    attn_v_masks_logits_kli = ablated_mask_logits_dict_edge[f'blocks.{k}.edge_mask_attention_v_logits'][:,l][1:][layer_i*13:(layer_i+1)*13]
                    
                    # print(attn_q_masks_logits_kli.shape)
                    
                    attn_q_masks_logits_kli[head_j] = -100.
                    attn_k_masks_logits_kli[head_j] = -100.
                    attn_v_masks_logits_kli[head_j] = -100.
    
        ablated_mask_logits_dict_edge['edge_mask_output_logits'][1:][layer_i*13:(1+layer_i)*13][head_j] = -100.
            
    return ablated_mask_logits_dict_edge
```

```python
# load IOI data
# Note that there are overlaps between train and test sets, due to the way IOIDataset is constructed (randomly sample N items)
ioi_prompts = pickle.load(open(join(data_dir, f'ioi_prompts_{ds_idx}.p'), 'rb'))
full_model_target_log_probs = torch.load(f'full_model_results/target_log_probs_{ds_idx}.pt')
full_model_pred_labels = torch.load(f'full_model_results/pred_labels_{ds_idx}.pt')

data_dict = prepare_ioi_data_for_clm(ioi_prompts, full_model_target_log_probs, full_model_pred_labels)
ds = IOICircuitDataset(data_dict)
dl = DataLoader(
    ds,
    batch_size=32
)
```

```python
acc_drops = {
    'djp-e': [],
    'in-the-wild': [],
}
```

```python
# ours -- edge-only

head_idx_by_layer_djp = [
    list(range(1, 11)),
    list(range(1, 12)),
    [2,7,9,11],
    [0,5,10],
    [8],
    [9],
    [6],
    [0],
    [10],
    [0],
    [7],
    [0,8,10]
]


for i in tqdm(range(12)):
    head_idx_i = head_idx_by_layer_djp[i]
    
    # circuit_gpt initialization
    model_path = join(model_dir, model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    tokenizer.pad_token = tokenizer.eos_token
    
    device = torch.device('cuda')    
    gpt_weights = torch.load(join(model_path, 'model_weights.pt')) 

    # head ablation
    mask_logits_dict_edge = torch.load(
        join(mask_dir, f'mask_logits_dict_edge_ioi_{ds_idx}_weight_0_edge_{n_epoch_e}.pt')
    )
    mask_logits_dict_edge = {k:v.detach().cpu() for k,v in mask_logits_dict_edge.items()}
    torch.cuda.empty_cache()
    ablated_mask_logits_dict_edge = head_ablation(mask_logits_dict_edge, [i]*len(head_idx_i), head_idx_i)
            
    circuit_gpt_config = CircuitGPTConfig(
        debug=False,
        gs_temp_weight=1.0,
        gs_temp_edge=1.0,
        use_weight_masks=False,
        use_edge_masks=True
    )
    circuit_gpt = CircuitGPT(circuit_gpt_config)
    
    circuit_gpt.load_pretrained_weight(gpt_weights)   

    # ablated_mask_logits_dict_edge = head_ablation(circuit_gpt.mask_logits_dict_edge, [i]*len(head_idx_i), head_idx_i)
    circuit_gpt.load_pretrained_edge_mask(ablated_mask_logits_dict_edge)
    circuit_gpt.to(device);   
    
    #################
    
    eval_results_ablated_model = eval_model(circuit_gpt, dl, tokenizer, device, 
        use_weight_mask=False, use_edge_mask=True, reverse=False
    )
    # print(
    #     f"Epoch 0. mean pruned model eval accuracy: {eval_results_ablated_model['acc']:.6f}," + 
    #     f"mean eval kl-div: {eval_results_ablated_model['kl']:.4f}," + 
    #     f"weight density: {eval_results_ablated_model['weight_density']:.4f}," + 
    #     f"edge density: {eval_results_ablated_model['edge_density']:.4f}"
    # )
    acc_drops['djp-e'].append(1.0 - eval_results_ablated_model['acc'])
    # print(acc_drops['djp-e'][-1])
    
    torch.cuda.empty_cache()
```

```python
# in-the-wild
# acc_drops['in-the-wild'] = []

head_idx_by_layer_itw = [
    [1,10],
    [0],
    [2],
    [0],
    [11],
    [5,8,9],
    [9],
    [3,9],
    [6,10],
    [0,6,7,9],
    [0,1,2,6,7,10],
    [2,9,10]
]


for i in tqdm(range(12)):
    head_idx_i = head_idx_by_layer_itw[i]
        
    # circuit_gpt initialization
    model_path = join(model_dir, model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    tokenizer.pad_token = tokenizer.eos_token
    
    device = torch.device('cuda')    
    gpt_weights = torch.load(join(model_path, 'model_weights.pt')) 
            
    circuit_gpt_config = CircuitGPTConfig(
        debug=False,
        gs_temp_weight=1.0,
        gs_temp_edge=1.0,
        use_weight_masks=False,
        use_edge_masks=True
    )
    circuit_gpt = CircuitGPT(circuit_gpt_config)
    
    circuit_gpt.load_pretrained_weight(gpt_weights)   

    ablated_mask_logits_dict_edge = head_ablation(circuit_gpt.mask_logits_dict_edge, [i]*len(head_idx_i), head_idx_i)
    circuit_gpt.load_pretrained_edge_mask(ablated_mask_logits_dict_edge)
    circuit_gpt.to(device);   
    
    #################
    
    eval_results_ablated_model = eval_model(circuit_gpt, dl, tokenizer, device, 
        use_weight_mask=False, use_edge_mask=True, reverse=False
    )
    # print(
    #     f"Epoch 0. mean pruned model eval accuracy: {eval_results_ablated_model['acc']:.6f}," + 
    #     f"mean eval kl-div: {eval_results_ablated_model['kl']:.4f}," + 
    #     f"weight density: {eval_results_ablated_model['weight_density']:.4f}," + 
    #     f"edge density: {eval_results_ablated_model['edge_density']:.4f}"
    # )
    acc_drops['in-the-wild'].append(1.0 - eval_results_ablated_model['acc'])
    # print(acc_drops['in-the-wild'][-1])
    torch.cuda.empty_cache()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
sns.set_theme()

fig, ax = plt.subplots(1,3, figsize=(18,3.2))

hm_djp = np.zeros((12,12))
hm_itw = np.zeros((12,12))

for i in range(12):
    for j in head_idx_by_layer_djp[i]:
        hm_djp[i][j] = 1.
    for j in head_idx_by_layer_itw[i]:    
        hm_itw[i][j] = 1.

# cmap_dict_djp = {0: '#FFFFFF', 1: '#ff2a00'}
# cmap_djp = ListedColormap([cmap_dict_djp[i] for i in range(2)])
# cmap_dict_itw = {0: '#FFFFFF', 1: '#ff2a00'}
# cmap_itw = ListedColormap([cmap_dict_itw[i] for i in range(2)])

sns.color_palette("light:b", as_cmap=True)
sns.heatmap(hm_djp, ax=ax[0], lw=1, cbar=False, mask=(hm_djp==0), vmin=0, vmax=1, cmap=sns.color_palette("Blues"))

sns.color_palette("light:r", as_cmap=True)
sns.heatmap(hm_itw, ax=ax[1], lw=1, cbar=False, mask=(hm_itw==0), vmin=0, vmax=1, cmap=sns.color_palette("Reds"))

ax[0].invert_yaxis(); ax[1].invert_yaxis()
ax[0].grid(False); ax[1].grid(False)

ax[0].set_xlabel('Head idx'); ax[1].set_xlabel('Head idx')
ax[0].set_ylabel('Layer'); ax[1].set_ylabel('Layer')
ax[2].set_ylabel('Acc. Drop'); ax[2].set_xlabel('Layer')

ax[0].set_title('Attn Heads: Disco-JP')
ax[1].set_title('Attn Heads: In-The-Wild')
ax[2].set_title('Layerwise Head Importance')

ax[2].plot(acc_drops['djp-e'], label='Disco-JP', color='blue')
ax[2].plot(acc_drops['in-the-wild'], label='In-The-Wild',  color='red')

ax[2].set_xticks(range(12));
ax[2].legend()

plt.savefig('figures/ioi-head-analysis.pdf', bbox_inches='tight')
```
