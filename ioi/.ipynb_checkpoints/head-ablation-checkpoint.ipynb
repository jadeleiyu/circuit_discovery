```python
from transformers import AutoTokenizer, AutoModelForCausalLM
hf_model = AutoModelForCausalLM.from_pretrained(join(model_dir, 'gpt2'))
tokenizer = AutoTokenizer.from_pretrained(join(model_dir, 'gpt2'))
```

```python
hooked_model = transformer_lens.HookedTransformer.from_pretrained(
    model_name='gpt2-small', hf_model=hf_model, 
    tokenizer=tokenizer
).to('cuda')
```

```python
def eval_with_head_ablation(hooked_model, data_dict, layer_to_ablate, head_index_to_ablate):
    
    def head_ablation_hook(value, hook):
        value[:, :, head_index_to_ablate, :] = 0.
        return value

    n_correct = 0
    N = len(data_dict['targets'])
    for prompt, (correct_answer, incorrect_answer) in tqdm(zip(data_dict['prompt'], data_dict['targets']), total=N):
        input_tokens = hooked_model.to_tokens(prompt)
        correct_index = hooked_model.to_single_token(' '+correct_answer)
        incorrect_index = hooked_model.to_single_token(' '+incorrect_answer)
    
        ablated_logits = hooked_model.run_with_hooks(
            input_tokens, 
            return_type="logits", 
            fwd_hooks=[(
                utils.get_act_name("v", layer_to_ablate), 
                head_ablation_hook
                )]
        )
        if ablated_logits[0,-1,correct_index] > ablated_logits[0,-1,incorrect_index]:
            n_correct += 1
    return n_correct
```

```python
ioi_prompts = pickle.load(open(join(data_dir, f'ioi_prompts_0.p'), 'rb'))
full_model_target_log_probs = torch.load(f'full_model_results/target_log_probs_0.pt')
full_model_pred_labels = torch.load(f'full_model_results/pred_labels_0.pt')

data_dict = prepare_ioi_data_for_clm(ioi_prompts, full_model_target_log_probs, full_model_pred_labels)
```

```python
for i in range(12):
    for j in range(12):
        n_correct = eval_with_head_ablation(hooked_model, data_dict, i, j)
        print(f'n_correct after ablating attn head {i}-{j}: {n_correct}')
```
